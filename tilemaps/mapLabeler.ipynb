```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'map1 source.png'

# Load image into np array with alpha channel
image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

# Add grid lines
for i in range(16):
    image[i::16, 0::16] = [0, 0, 255, 255]
    image[0::16, i::16] = [0, 0, 255, 255]

for i in range(11):
    o = 160 * i * 2
    # Add black top for first 160x16 pixels
    image[0:16, o:o+160] = [0, 0, 0, 255]
    # Add white left for first 16x160 pixels
    image[0:16, o + 160:o + 2*160] = [255, 255, 255, 255]

# Add a green line 
image[16*10,:] = [0, 255, 0, 255]
image[16*(10-5),:] = [0, 255, 0, 255]

# Add Working Line
o = 16*10
image[0:16, o:o+160] = [0, 255, 255, 255]

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

# Save image
cv2.imwrite('labeledMap1.png', image)
```
